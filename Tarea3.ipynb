#Gemini Integrado
Copia y pega este prompt en el apartado de generar codigo con Gemini. *Prompt:* **Genera un serie de datos aleatorios de estaturas de personas. Calcula media desviación estándar y has un gráfico de frecuencias**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios de estaturas (simulando una distribución normal)
# Media: 170 cm, Desviación estándar: 10 cm, Número de personas: 1000
estaturas = np.random.normal(loc=170, scale=10, size=1000)

# Calcular la media y la desviación estándar
media_estaturas = np.mean(estaturas)
desviacion_estandar_estaturas = np.std(estaturas)

print(f"Media de las estaturas: {media_estaturas:.2f} cm")
print(f"Desviación estándar de las estaturas: {desviacion_estandar_estaturas:.2f} cm")

# Crear un gráfico de frecuencias (histograma)
plt.figure(figsize=(10, 6))
plt.hist(estaturas, bins=30, edgecolor='black')
plt.title('Histograma de Frecuencias de Estaturas')
plt.xlabel('Estatura (cm)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

# Ejemplo y tarea:


1.   Crea un data Frame con los datos de la Hoja de Excel "Notas_Laboratorio.xlsx". Las columnas tienen encabezados.
2.   Separa el nombre y el apellido en distintas columnas. Por ejemplo "Kip Thorne" se debe separa en "Kip" en la columna Nombre y "Thorne" en la columna Apellido. Asegurate que estas columnas se coloquen al principio y se elimine la columna original.
3. Calcula la Nota final Promediando las notas desde P1 hasta P10.
4. La nota de aprobación es 14. Pinta de naranja los alumnos con nota menor a 14. Con gris los que tienen entre 14 y 17 y en verde los que tienen nota superior.
5. Has un histograma de 40 bines con el criterio que hasta 14 este coloreado de naranja hasta 17 de gris y el resto en verde



```python
import pandas as pd

# Define the path to the Excel file in Google Drive
# Make sure the path is correct based on where you mounted your drive and the file location
excel_file_path = "/content/drive/My Drive/Colab Notebooks/cuantitativa/Notas_Laboratorio.xlsx"

# Read the Excel file into a pandas DataFrame
# The sheet name is 'Hoja1' by default, but if it's different, you'll need to specify it
try:
    df_notas = pd.read_excel(excel_file_path)
    print("DataFrame created successfully:")
    display(df_notas.head())
except FileNotFoundError:
    print(f"Error: The file was not found at {excel_file_path}")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
display(df_notas)
```

```python
# 2. Separate Nombre and Apellido
# Split the 'Nombre Apellido' column into two new columns 'Nombre' and 'Apellido'
split_names = df_notas['Nombre Apellido'].str.split(' ', n=1, expand=True)

# Insert the new 'Nombre' and 'Apellido' columns at the beginning of the DataFrame
df_notas.insert(0, 'Apellido', split_names[1])
df_notas.insert(0, 'Nombre', split_names[0])

# Drop the original 'Nombre Apellido' column
df_notas = df_notas.drop('Nombre Apellido', axis=1)

print("DataFrame after splitting names:")
display(df_notas.head())
```

```python
# 3. Calculate the final grade
# Select the columns containing the grades (P1 to P10)
grade_columns = [f'P{i}' for i in range(1, 11)]

# Calculate the average of these columns for each row and create a new column 'Nota final'
df_notas['Nota final'] = df_notas[grade_columns].mean(axis=1)

print("DataFrame with 'Nota final' column:")
display(df_notas.head())
```

```python
# 4. Apply conditional formatting based on final grade
def color_grades(val):
    if isinstance(val, (int, float)):
        if val < 14:
            return 'background-color: darkorange'
        elif 14 <= val <= 17:
            return 'background-color: grey'
        else:
            return 'background-color: green'
    return ''

# Apply the styling and format the 'Nota final' column
styled_df = df_notas.style.map(color_grades, subset=['Nota final']).format({'Nota final': '{:.2f}'})

print("DataFrame with conditional formatting:")
display(styled_df)
```

```python
import matplotlib.pyplot as plt

# 5. Create a histogram with colored bars based on grade ranges
plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(df_notas['Nota final'], bins=40, edgecolor='black')

# Define the colors based on the grade ranges
for patch, bin_edge in zip(patches, bins):
    if bin_edge < 14:
        patch.set_facecolor('orange')
    elif 14 <= bin_edge <= 17:
        patch.set_facecolor('lightgrey')
    else:
        patch.set_facecolor('lightgreen')

plt.title('Histogram of Final Grades with Approval Criteria')
plt.xlabel('Nota final')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```
